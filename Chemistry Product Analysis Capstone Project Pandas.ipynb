```python
#Terpene averages analysis

terpeneDF = chem[["Terpinolene....","Eudesmol....","α.Pinene....","β.Pinene....","Camphene....",
            "Eucalyptol....","Isopulegol....","Menthol....","Borneol....","Trans.Beta.Ocimene....",
            "Cis.Beta.Ocimene....","Geraniol....","Linalool....","Guaiol....","α.Bisabolol....","α.Terpinene....",
            "Citronellol....","δ.3.Carene....","Cis.Nerolidol....","Trans.Nerolidol....", "Limonene....","P.Cymene....",
            "γ.Terpinene....","α.Terpineol....","ß.Myrcene....", "CaryophylleneOxide....","α.Humulene....",
            "ß.Caryophyllene...."]]

# print(terpeneDF)
for terp in terpeneDF:
    terpeneDF[terp] = pd.to_numeric(terpeneDF[terp], errors='coerce')
    print(f"The Average mg/g concentration of {terp} : {terpeneDF[terp].mean(skipna=True)}")

print("\n")
    
#Returns the value of the lowest mean in the dataframe
lowest_mean = terpeneDF.mean().min()
# Returns the name of the column that has the lowest value in the dataframe
lowest_mean_column = terpeneDF.mean().idxmin()
print(f"Lowest Mean Value: {lowest_mean} in column: {lowest_mean_column}")

print("\n")

#Returns the value of the highest mean in the dataframe
highest_mean = terpeneDF.mean().max()
# Returns the name of the column that has the highest value in the dataframe
highest_mean_column = terpeneDF.mean().idxmax()
print(f"Highest Mean Value: {highest_mean} in column: {highest_mean_column}")
```

```python
# Bar graph representing the mean concentrations of each individual terpene. 
concentrations = terpeneDF.mean()

# Create the bar plot
plt.bar(concentrations.index, concentrations.values)
plt.xticks(rotation=90)
plt.title("Terpene Concentrations")
plt.xlabel("Terpene")
plt.ylabel("Concentration (mg/g)")

# Display the plot
plt.show()
```
